3. Pregunta 3
    - CSV Global: "Country name" "Year" "Life Ladder"
    - CSV 2021: "Country Name" "Year"(que no existe y deberemos crearlo) "Ladder Score"
4. Pregunta 4

    - CSV 2021: "Country Name" "Year"(que no existe y deberemos crearlo) "Logged GDP per capita"

    - CSV 2021: "Year"(que no existe y deberemos crearlo)"Logged GDP per capita"

Por ello, el dataFrame maestro contará con las siguientes columnas:
- Country
- Year
- Ladder
- GDP
- Life expectancy
- Continent

Una vez creado el dataFrame, procederemos a responder a cada pregunta


### Filtrado de CSV general

```python
import pandas as pd
```

```python
dataFrame1 = pd.read_csv('world-happiness-report.csv')
dataFrame1.head(5)
```

```python
filtered_dataFrame1 = dataFrame1.drop(columns=['Social support', 'Freedom to make life choices','Generosity','Perceptions of corruption','Positive affect','Negative affect']).rename(columns = {'Country name':'Country','year':'Year','Life Ladder':'Ladder','Log GDP per capita':'GDP','Healthy life expectancy at birth':'Life expectancy'})
filtered_dataFrame1.head(5)
```

### Filtrado de CSV 2021

```python
dataFrame2 = pd.read_csv('world-happiness-report-2021.csv')
dataFrame2.head(5)
```

```python
filtered_dataFrame2 = dataFrame2[['Country name','Regional indicator','Ladder score','Logged GDP per capita','Healthy life expectancy']].rename(columns = {'Country name':'Country','Regional indicator':'Continent','Ladder score':'Ladder','Logged GDP per capita':'GDP','Healthy life expectancy':'Life expectancy'})
filtered_dataFrame2.insert(1,'Year',2021)
filtered_dataFrame2.head(5)
```

### Creación de CSV maestro

```python
master_dataFrame = pd.concat([filtered_dataFrame1,filtered_dataFrame2]).sort_values(by=['Country','Year']).reset_index(drop=True)
master_dataFrame
```

### Pregunta 1
¿Cuál es el país más “feliz” del 2021 según la data? (considerar que la columna “Ladder score” mayor número más feliz es el país)

```python
df1 = master_dataFrame.query('Year == 2021').nlargest(1,'Ladder')
df1[['Continent','Country','Year','Ladder']]
```

### Pregunta 
¿Cuál es el país más “feliz” del 2021 por continente según la data?

```python
df_2021 = master_dataFrame.query('Year == 2021')
happiest_index = df_2021.groupby('Continent')['Ladder'].idxmax()
countries = df_2021.loc[happiest_index]
countries[['Continent','Country','Ladder']].sort_values(by=['Ladder'])
```

### Pregunta 3
¿Cuál es el país que más veces ocupó el primer lugar en todos los años?

```python
ranking = master_dataFrame.groupby(['Year'])['Ladder'].idxmax()
df3 = master_dataFrame.loc[ranking].sort_values(by =['Year'])
df_country = df3.groupby('Country').count().sort_values(by=['Year'],ascending=False).head(1).Year
df_ganador = pd.DataFrame(df_country,[df_country.index[0]],['Year']).rename(columns = {'Year':'Numero_veces_primero'})
df_ganador
```

### Pregunta 4
¿Qué puesto de Felicidad tiene el país con mayor GDP del 2020?

```python
df3 = master_dataFrame.query('Year == 2020').sort_values(by =['GDP'], ascending=False)
df3['rank'] = df3['Ladder'].rank()
df3[['Country','GDP','rank']].head(1)

```

### Pregunta 5
¿En que porcentaje a variado a nivel mundial el GDP promedio del 2020 respecto al 2021? ¿Aumentó o disminuyó?

```python
prom_2020 = master_dataFrame.query('Year == 2020').GDP.mean()
prom_2021 = master_dataFrame.query('Year == 2021').GDP.mean()
variance = ((prom_2021 - prom_2020) / prom_2020) * 100
variance
```

### Pregunta 6
¿Cuál es el país con mayor expectativa de vida (“Healthy life expectancy at birth”)? Y ¿Cuánto tenia en ese indicador en el 2019?

```python

index = master_dataFrame.query("Year == 2021")["Life expectancy"].idxmax()
country = master_dataFrame.iloc[index][["Country", "Year", "Life expectancy"]]
print("El pais con más esperanza de vida es: " + country["Country"])
```

```python
print("Laexpectativa de " + country["Country"] + " en el 2019 es:")
master_dataFrame[master_dataFrame.Country == country.Country].query("Year == 2019")[["Country", "Year", "Life expectancy"]]
```

```python
print("La expectativa de " + country["Country"] + " en el año 2020 es:")
master_dataFrame[master_dataFrame.Country == country.Country].query("Year == 2020")[["Country", "Year", "Life expectancy"]]
```
